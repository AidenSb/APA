```r
library(Seurat)
library(ggplot2)
library(dplyr)
library(stringr)
library(Sierra)
library(presto)
library(ggplot2)
library(msigdbr)
library(fgsea)
library(sctransform)
library(tidyverse)
setwd('/data/APAproject/post_qual/data/control_vs_control/')
```

```r
reference.file <- '/home/aiden/data/refgenome/refdata-gex-GRCh38-2020-A/genes//genes.gtf'
gtf_gr <- rtracklayer::import(reference.file)
gtf_TxDb <- GenomicFeatures::makeTxDbFromGFF(reference.file, format="gtf")
```

```r
# peak.annotations <- read.table("sierra_run/CTdim_CTnagy_peaks_annotated.txt", 
#                                header = TRUE,
#                                sep = "\t",
#                                row.names = 1,
#                                stringsAsFactors = FALSE)
peak.counts <- ReadPeakCounts(data.dir = "sierra_run/CTdim_CTnagy_Sierra_aggregate/")
CT_all_peaks_so <- NewPeakSeurat(peak.data = peak.counts, 
                              annot.info = peak.annotations, 
                              min.cells = 50, min.peaks = 200)
CT_all_peaks_so
```

```r
## metadata for each group
## matadata for control cells
ct_dim <- readRDS('../Dimitry_Velmeshev_ASD/rdata/Control_cell_types_GE_PFC.rds')
ct_dim_meta_data <- data.frame(clusters=ct_dim$cluster, 
                           main_celltype=paste0('Dim_',ct_dim$main_cellTypes))
ct_dim_meta_data$cond <- 'Dim'
ct_dim_meta_data$cells <- rownames(ct_dim_meta_data)
ct_dim_meta_data
```

```r
ct_nagy <- readRDS('../Nagy_mdd/rdata/ge_so_control_Nagy.rds')
```

```r
ct_nagy$main_cellTypes <- 'NA'
IN_idx <- which(ct_nagy$cluster %in% c('Inhibitory_SST',' Inhibitory_VIP','Inhibitory_VIP',
                                     'Inhibitory','Inhibitory_PVALB'))
ct_nagy$main_cellTypes[IN_idx] = 'Inhibitory_Neurons'
##
Ex_idx <- which(ct_nagy$cluster %in% c('Excitatory_L4-5','Excitatory_L5-6','Excitatory_L4-6',
                                     'Excitatory_L5','Excitatory_L2-4'))
ct_nagy$main_cellTypes[Ex_idx] = 'Excitatory_Neurons'
##
En_idx <- which(ct_nagy$cluster %in% c('Endothelial'))
ct_nagy$main_cellTypes[En_idx] = 'Endothelial'
##
Ast_idx <- which(ct_nagy$cluster %in% c('Astrocyte'))
ct_nagy$main_cellTypes[Ast_idx] = 'Astrocytes'
##
Mg_idx <- which(ct_nagy$cluster %in% c('Microglia/Macrophages'))
ct_nagy$main_cellTypes[Mg_idx] = 'Microglia'
##
Og_idx <- which(ct_nagy$cluster %in% c('Oligodendrocytes'))
ct_nagy$main_cellTypes[Og_idx] = 'Oligodendrocytes'

##
OPC_idx <- which(ct_nagy$cluster %in% c('OPCs'))
ct_nagy$main_cellTypes[OPC_idx] = 'OPC'
```

```r
ct_nagy_meta_data <- data.frame(clusters=ct_nagy$cluster, 
                           main_celltype=paste0('Nagy_',ct_nagy$main_cellTypes))
ct_nagy_meta_data$cond <- 'Nagy'
ct_nagy_meta_data$cells <- rownames(ct_nagy_meta_data)
ct_nagy_meta_data
```

```r
Control_metadata <- rbind(ct_dim_meta_data,ct_nagy_meta_data)
Control_metadata
```

```r
write.table(Control_metadata, file = 'rdata/dim_nagy_control_metadata.tsv', sep='\t', quote=F)
```

```r
CT_all_peaks_so$cellnames <- Cells(CT_all_peaks_so)
CT_all_peaks_so_filtered <- subset(CT_all_peaks_so, subset= cellnames %in% Control_metadata$cells)
CT_all_peaks_so_filtered <- AddMetaData(CT_all_peaks_so_filtered, metadata = Control_metadata)
CT_all_peaks_so_filtered
```

```r
saveRDS(CT_all_peaks_so_filtered, 'rdata/all_control_dim_and_nagy_peaks_so.rds')
```

```r
CT_all_peaks_so_filtered <- SetIdent(CT_all_peaks_so_filtered, 
                                    value= CT_all_peaks_so_filtered$main_celltype)
```
