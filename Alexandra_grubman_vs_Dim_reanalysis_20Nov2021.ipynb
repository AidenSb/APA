```r
calculate_lfc_CTvsAD <- function(ct_so, ad_so, features){
    ad_rbp <- as.data.frame(AverageExpression(ad_so, features = features,assays = 'SCT' ))
    ct_rbp <- as.data.frame(AverageExpression(ct_so, features = features,assays = 'SCT' ))
    rbp_lfc <- as.data.frame(log(as.numeric(ct_rbp$all)/as.numeric(ad_rbp$all), base = 2))
    rownames(rbp_lfc) <-  rownames(ct_rbp)
    return(rbp_lfc)
}
```
