## rename the SRR files with regard to patient group

```python
os.chdir('/data/APAproject/post_qual/data/Alexandra_Grubman/Raw_and_pre_processed/')
os.listdir()
```

```python
tmp = pd.read_csv('download_data.tsv', sep=' ', header=None)
sample = {}
for f in tmp[4]:
    sampname = f.split('_')[0]
    patientgroup = f.split('_')[-4]
    sample[sampname] = patientgroup
sample
```

```python
all_dirs = [f for f in os.listdir() if ( "_out" in f and ".mro" not in f)]
outfile = open('bamfiles.txt', 'w+')
for directory in all_dirs:
    bamin = directory+ "/outs/possorted_genome_bam.bam"
    patgroup = directory.split("_")[0]+"_"+sample[directory.split("_")[0]]
    bamout = bamin.replace("possorted_genome_bam", patgroup)
#     cmd = 'mv {} {}'.format(bamin, bamout)
#     print(cmd)
#     subprocess.call(cmd,shell=True)
    bamiout = bamout + '.bai'
    bamiin = bamin + '.bai'
    cmd = 'mv {} {}'.format(bamiin,bamiout)
    subprocess.call(cmd,shell=True)
    outfile.write(bamout+"\n")
outfile.close()
```

## run the regtools on the bamfiles

```python
%% bash 
while read l; do out=${l/.bam/.bed}; regtools junctions extract -s 1 $l -o $out; done < bamfiles.txt 
```

## Make the metadata file for Sierra pipeline

```python
bamfiles = pd.read_csv('../Raw_and_pre_processed/bamfiles.txt', header=None)
barcodes_list, names_list, junctions_list = [], [], []
for i,row in bamfiles.iterrows():
    bam = row[0]
    junction = bam.replace('.bam','.bed')
    junctions_list.append(junction)
    bam_name = bam.split('/')[-1]
    barcode = bam.split('/')[:-1]
    barcode.append('filtered_feature_bc_matrix/barcodes.tsv')
    barcode = "/".join(barcode)
    barcodes_list.append(barcode)
    name = bam_name.split('.')[0]
    names_list.append(name)
bamfiles['junction'] = junctions_list
bamfiles['barcodes'] = barcodes_list
bamfiles['name'] = names_list
bamfiles.columns = ['bam','junction','barcodes','name']
bamfiles.to_csv('metadata_for_sierra.tsv', sep='\t', index=False)
```
