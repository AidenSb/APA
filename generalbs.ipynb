```python
function (peaks.object, gtf_gr, gtf_TxDb, population.1, population.2 = NULL, 
    exp.thresh = 0.1, fc.thresh = 0.25, adj.pval.thresh = 0.05, 
    num.splits = 6, seed.use = 1, verbose = TRUE, do.MAPlot = FALSE, 
    ncores = 1) 
{
    res.table <- DUTest(peaks.object = peaks.object, population.1 = population.1, 
        population.2 = population.2, exp.thresh = exp.thresh, 
        fc.thresh = fc.thresh, adj.pval.thresh = adj.pval.thresh, 
        num.splits = num.splits, seed.use = seed.use, feature.type = c("UTR3"), 
        verbose = verbose, do.MAPlot = do.MAPlot, ncores = ncores)
    if (verbose) 
        print("Filtering for alternative transcript usage")
    all.peaks = rownames(res.table)
    strand = sub(".*:.*:.*-.*:(.*)", "\\1", all.peaks)
    strand = plyr::mapvalues(x = strand, from = c("1", "-1"), 
        to = c("+", "-"))
    peak.remainder = sub(".*:(.*:.*-.*):.*", "\\1", all.peaks)
    peaks.use = paste0(peak.remainder, ":", strand)
    res.table$granges_peaks <- peaks.use
    du.peaks.gr <- GenomicRanges::GRanges(peaks.use)
    peaks.expressed <- GetExpressedPeaks(peaks.object, population.1 = population.1, 
        population.2 = population.2, threshold = exp.thresh)
    genes.expressed <- sub("(.*):.*:.*-.*:.*", "\\1", peaks.expressed)
    peaks.use.idx <- which(genes.expressed %in% as.character(res.table$gene_name))
    peaks.expressed <- peaks.expressed[peaks.use.idx]
    strand = sub(".*:.*:.*-.*:(.*)", "\\1", peaks.expressed)
    strand = plyr::mapvalues(x = strand, from = c("1", "-1"), 
        to = c("+", "-"))
    peak.remainder = sub(".*:(.*:.*-.*):.*", "\\1", peaks.expressed)
    peaks.expressed.granges = paste0(peak.remainder, ":", strand)
    duplicate.peaks <- names(table(peaks.expressed.granges))[which(table(peaks.expressed.granges) > 
        1)]
    if (length(duplicate.peaks) > 0) {
        peaks.remove.idx <- which(peaks.expressed.granges %in% 
            duplicate.peaks)
        peaks.expressed.granges <- peaks.expressed.granges[-peaks.remove.idx]
        peaks.expressed <- peaks.expressed[-peaks.remove.idx]
        res.table <- res.table[intersect(rownames(res.table), 
            peaks.expressed)]
    }
    expressed.peaks.gr <- GenomicRanges::GRanges(peaks.expressed.granges)
    granges_peaks_mapping_table <- data.frame(PeakID = peaks.expressed, 
        row.names = peaks.expressed.granges, stringsAsFactors = FALSE)
    utr3.ref <- GenomicFeatures::threeUTRsByTranscript(gtf_TxDb)
    utr3.ref <- unlist(utr3.ref)
    all_UTR_3_hits <- GenomicRanges::findOverlaps(expressed.peaks.gr, 
        utr3.ref, type = "any")
    utr3.mappings <- as.data.frame(all_UTR_3_hits)
    query.hit.df <- as.data.frame(expressed.peaks.gr[utr3.mappings$queryHits, 
        ])
    subject.hit.df <- as.data.frame(utr3.ref[utr3.mappings$subjectHits, 
        ], row.names = as.character(1:nrow(utr3.mappings)))
    query.hit.df <- query.hit.df %>% dplyr::mutate(granges_peak = paste0(seqnames, 
        ":", start, "-", end, ":", strand))
    utr3.mappings$exon_name <- subject.hit.df$exon_name
    utr3.mappings$granges_peak <- query.hit.df$granges_peak
    peak.ids <- granges_peaks_mapping_table[as.character(query.hit.df$granges_peak), 
        "PeakID"]
    utr3.mappings$peakID <- peak.ids
    res.table$peak_name <- rownames(res.table)
    diff.transcript.check.values <- apply(as.matrix(res.table), 
        1, function(x) {
            diff.site <- x["peak_name"]
            this.gene <- x["gene_name"]
            all.sites <- SelectGenePeaks(peaks.object, gene = this.gene, 
                feature.type = "UTR3")
            sites.expressed <- intersect(all.sites, peaks.expressed)
            exons.diff <- unique(subset(utr3.mappings, peakID %in% 
                diff.site)$exon_name)
            remaining.sites <- setdiff(all.sites, diff.site)
            remaining.exons <- unique(subset(utr3.mappings, peakID %in% 
                remaining.sites)$exon_name)
            diff.transcript.check <- ifelse(length(setdiff(remaining.exons, 
                exons.diff)) > 0, TRUE, FALSE)
            return(diff.transcript.check)
        })
    res.table$Diff_transcript <- diff.transcript.check.values
    res.table <- subset(res.table, Diff_transcript == TRUE)
    peaks.to.annotate <- res.table$granges_peaks
    peaks.gr <- GenomicRanges::GRanges(peaks.to.annotate)
    granges_peaks_mapping_table <- data.frame(PeakID = rownames(res.table), 
        row.names = peaks.to.annotate, stringsAsFactors = FALSE)
    transcripts.ref <- GenomicFeatures::transcripts(gtf_TxDb)
    all_transcript_hits <- GenomicRanges::findOverlaps(peaks.gr, 
        transcripts.ref, type = "any")
    transcript.mappings <- as.data.frame(all_transcript_hits)
    query.hit.df <- as.data.frame(peaks.gr[transcript.mappings$queryHits, 
        ])
    subject.hit.df <- as.data.frame(transcripts.ref[transcript.mappings$subjectHits, 
        ], row.names = as.character(1:nrow(transcript.mappings)))
    query.hit.df <- query.hit.df %>% dplyr::mutate(granges_peak = paste0(seqnames, 
        ":", start, "-", end, ":", strand))
    transcript.mappings$transcript_name <- subject.hit.df$tx_name
    transcript.mappings$granges_peak <- query.hit.df$granges_peak
    peak.ids <- granges_peaks_mapping_table[as.character(query.hit.df$granges_peak), 
        "PeakID"]
    transcript.mappings$peakID <- peak.ids
    peak.transcript.table <- transcript.mappings %>% dplyr::group_by(peakID) %>% 
        dplyr::summarize(Transcript_names = paste(transcript_name, 
            collapse = ";")) %>% as.data.frame()
    rownames(peak.transcript.table) <- as.character(peak.transcript.table$peakID)
    peak.transcript.table <- peak.transcript.table[rownames(res.table), 
        ]
    res.table$Transcript_name <- as.character(peak.transcript.table$Transcript_names)
    res.table <- res.table[, c("gene_name", "genomic_feature(s)", 
        "population1_pct", "population2_pct", "pvalue", "padj", 
        "Log2_fold_change", "Transcript_name")]
    if (verbose) 
        print(paste0(nrow(res.table), " peaks detected as representing alternative transcript usage"))
    return(res.table)
}
```

```python
def clean_data(data) -> None:
    """Modify data so that the applicable string values are converted to their
    appropriate type.

    The indexes of the string values that are converted, and their
    corresponding type, are:
        - COLUMN_RIDING is an int
        - COLUMN_VOTER is an int
        - COLUMN_RANK is a list of strings
        - COLUMN_RANGE is a list of integers
        - COLUMN_APPROVAL is a list of booleans

    >>> row = ['0', '1', 'NDP;Liberal;Green;CPC', '1;4;2;3', 'NO;YES;NO;NO']
    >>> clean_data([row])
    >>> row == SAMPLE_DATA_1[0]
    True
    >>> row = ['117', '12', 'Liberal;CPC;NDP;Green', '4;0;5;0', 'YES;NO;YES;NO']
    >>> clean_data([row])
    >>> row == SAMPLE_DATA_2[0]
    True
    """
    for dat in data:    
        dat[COLUMN_RIDING] = int(dat[COLUMN_RIDING])
        dat[COLUMN_VOTER] = int(dat[COLUMN_VOTER])
        dat[COLUMN_RANK] = [str(rank).upper() for rank \
                            in dat[COLUMN_RANK].split(';')]
        dat[COLUMN_RANGE] = [int(rank) for rank in dat[COLUMN_RANGE].split(';')]
        dat[COLUMN_APPROVAL] = [(approval=='YES') for approval \
                                in dat[COLUMN_APPROVAL].split(';')]

```

```python
def voting_range(range_ballots):
    """Return a list of the total scores received by each party on 
    range_ballots. The list elements are ordered by PARTY_ORDER.

    >>> voting_range(extract_column(SAMPLE_DATA_1, COLUMN_RANGE))
    [9, 10, 10, 9]
    """
    lst = [0,0,0,0]
    for votes in range_ballots:
        for idx in range(len(votes)):
            lst[idx] += votes[idx]
    return lst
```

```python
def make_author_to_articles(x):
    """Return a dict that maps each author name to a list (sorted in
    lexicographic order) of IDs of articles written by that author,
    based on the information in id_to_article.

    >>> make_authorto_articles(EXAMPLE_ARXIV) == EXAMPLE_BY_AUTHOR
    True
    >>> make_author_to_articles({})
    {}
    """
    
    # Complete this Task 3 function
    author_article = {}
    for key in sorted(x.keys()):
        authors = x[key][AUTHORS]
        for author in authors:
            if author not in author_article:
                author_article[author] = []
                author_article[author].append(key)
            else:
                author_article[author].append(key)
    return author_article
```

```python
def has_prolific_authors(author_article,
                         article_ID, threshold ) -> bool:
    """ Checks if the author is prolific based on the threshold 
        number of articles (thrid parameter). returns True if and
        only if the article (second argument) has at least one
        author who is considered prolific

    >>> has_prolific_authors(EXAMPLE_BY_AUTHOR, '008', 2)
    True        
    """                         
    for author in author_article:
        if article_ID in author_article[author]:
            if len(author_article[author]) >= threshold:
                return True
    return False
```

```python
def rename_fastqs(files):
    checkmark, newnames = [], []
    newsample_checkmark = []
    i = 0
    for file in files:
        splited = file.split('_')
        samp_id = splited[0]
        biosamp_id = splited[1]
        if biosamp_id not in newsample_checkmark:
            i = 0
            newsample_checkmark.append(biosamp_id)
        if samp_id not in checkmark:
            checkmark.append(samp_id)
            i +=1 
            renamed = splited[2] + '_L00' + str(i) + "_R" + splited[-1].split(".")[0] + '_001.fastq.gz'
            newnames.append(renamed)
        else:
            renamed = splited[2] + '_L00' + str(i) +"_R" + splited[-1].split(".")[0] + '_001.fastq.gz'
            newnames.append(renamed)
    return newnames
```

```python
"""CSC 108 Final Assessment: Question 6 [10 marks]

This code is provided solely for the personal and private use of students
taking the CSC108 course at the University of Toronto. Copying for purposes
other than this use is expressly prohibited. All forms of distribution of
this code, whether as given or with any changes, are expressly prohibited.

This file is:
Copyright (c) 2021 Michelle Craig, Tom Fairgrieve, Sadia Sharmin, and
Jacqueline Smith.
"""

import copy
from typing import List

# Before your test time expires, upload this file to Markus as t3-q6.py

# In Assignment 2, you worked with VoteData.  Each voter's VoteData contains,
# in order, their riding number, their voter number, their Rank Ballot, their
# Range Ballot and their Approval Ballot.

# Here is a description of the VoteData type:
# VoteData = List[int, int, List[str], List[int], List[bool]]

# The Range Ballots on this test are slightly simplified from those in A2.
# For this question, Range Ballot values are ints from 1 to 4.
# They are given in PARTY_ORDER = ['CPC', 'GREEN', 'LIBERAL', 'NDP']. The
# higher the integer, the more this voter prefers this party. There are to
# be no duplicate values, so every Range Ballot contains the ints 1 to 4
# in some order.

# You may wish to use these constants in your code.
COLUMN_RANK = 2
COLUMN_RANGE = 3
PARTY_ORDER = ['CPC', 'GREEN', 'LIBERAL', 'NDP']

# PART 1: Complete the function make_range_from_rank below so that it
# works correctly as described in the docstring.


def make_range_from_rank(rank_ballot: List[str]) -> List[int]:
    """Return a range ballot that corresponds with the voter preferences in
    rank_ballot. The highest-ranked party gets 4 points, the second highest
    gets 3, the third highest gets 2, and the lowest-ranked party gets 1. The
    returned list of points per party is in the order specified in PARTY_ORDER.

    Precondition: len(rank_ballot) == 4, the four party names in rank_ballot
                  are unique and each appears in PARTY_ORDER

    >>> make_range_from_rank(['CPC', 'GREEN', 'NDP', 'LIBERAL'])
    [4, 3, 1, 2]
    >>> make_range_from_rank(['NDP', 'LIBERAL', 'GREEN', 'CPC'])
    [1, 2, 3, 4]
    """

    # TODO: Write your code here
    new_range = [0,0,0,0]
    i = 4
    for vote in rank_ballot:
        idx = PARTY_ORDER.index(vote)
        new_range[idx] = i
        i -= 1
    return new_range
# PART 2: We consider a voter's Range Ballot to be consistent with their Rank
# Ballot if and only if their Range Ballot is the same as the ballot that would
# be generated from their Rank Ballot.

# Complete the function below according to its docstring. Your solution MUST
# use the function make_range_from_rank as a helper function.

# Even if you are unable to fully answer PART 1 above, you should write
# your code for PART 2 as if your make_range_from_rank function is correct.


# Sample data for examples
SAMPLE_DATA = [[0, 1, ['NDP', 'LIBERAL', 'CPC', 'GREEN'], [1, 2, 3, 4],
                [False, True, False, False]],
               [1, 2, ['CPC', 'GREEN', 'NDP', 'LIBERAL'], [4, 3, 1, 2],
                [False, False, True, True]],
               [1, 3, ['CPC', 'GREEN', 'NDP', 'LIBERAL'], [1, 3, 4, 2],
                [False, True, False, True]]]

CORRECTED = [[0, 1, ['NDP', 'LIBERAL', 'CPC', 'GREEN'], [2, 1, 3, 4],
              [False, True, False, False]],
             [1, 2, ['CPC', 'GREEN', 'NDP', 'LIBERAL'], [4, 3, 1, 2],
              [False, False, True, True]],
             [1, 3, ['CPC', 'GREEN', 'NDP', 'LIBERAL'], [4, 3, 1, 2],
              [False, True, False, True]]]


def replace_inconsistent(all_votes: List['VoteData']) -> int:
    """Modify the VoteData of every voter in all_votes that has a Range Ballot
    that is NOT consistent with their Rank Ballot by replacing their Range
    Ballot with a new, consistent, ballot that is based on their Rank Ballot.

    Do NOT modify any VoteData where the given Range Ballot is consistent
    with the given Rank Ballot.

    Return the number of voters whose Range Ballot has been modified.

    >>> data1 = copy.deepcopy(SAMPLE_DATA)
    >>> replace_inconsistent(data1)
    2
    >>> data1 == CORRECTED
    True
    """

    # TODO: Write your code here
    count = 0
    for vote in all_votes:
        ranked_ballot = vote[COLUMN_RANK]
        range_ballot = vote[COLUMN_RANGE]
        new_range = make_range_from_rank(ranked_ballot)
        if new_range != range_ballot:
            vote[COLUMN_RANGE] = new_range
            count += 1
    return count
```
