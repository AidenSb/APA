## dist plotting the REDU for MAAPER outputs

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
os.chdir('/data1/APA/Paul_ALS_Data/bams_in/subscelltype_bamfiles/Mapper_outs/')
```

```python
CTs = [ct for ct in os.listdir() if 'CTRL' in ct]
CTs = list(set(CTs))
CTs
```

```python
exc_df = pd.read_csv('Excitatory_C9ALSvsCTRL/gene.txt', sep='\t')
exc_df
```

```python
def plot_redu(df, name):
    dat = df['REDu']
    dat.replace([np.inf, -np.inf], np.nan, inplace=True)
    dat.dropna(inplace=True)
    ax = sns.distplot(dat)
    ax.set(xlabel='REDu', ylabel='Density')
    plt.axvline(dat.median(),
                color='red', label=round(dat.median(),2))
    plt.legend()
    outname = "REDU_plots/" + name + '_REDu_distplots.png'
    plt.savefig(outname)
    plt.clf()
```

```python
for ct_cn in CTs:
    file = ct_cn + "/gene.txt"
    df = pd.read_csv(file, sep='\t')
    print(ct_cn)
    print(df.shape)
    plot_redu(df, ct_cn)
```
