```r
df_redu = read.csv("/data1/wgs_APA/results/JW/cpa_df_redu.csv", header = T, row.names = 1)
df_redi = read.csv("/data1/wgs_APA/results/JW/cpa_df_redi.csv", header = T, row.names = 1)
df_redu_corr = read.csv("/data1/wgs_APA/results/JW/cpa_df_redu_corr1.csv", header = T, row.names = 1)
df_redi_corr = read.csv("/data1/wgs_APA/results/JW/cpa_df_redu_corr2.csv", header = T, row.names = 1)
df_stratified = read.csv("/data1/wgs_APA/results/JW/JW_stratification_df.csv", header = T, row.names = 1)

df_redu
```

```r
# plot correlation heatmap using pheatmap for df_redu_corr which is already the corr df
pheatmap(df_redu_corr)
```

```r
# plot  df_redu_corr and  df_redi_corr side by side
par(mfrow=c(1,2))
p <- pheatmap(df_redu_corr)
p <- pheatmap(df_redi_corr)
p
```

```r
p <- pheatmap(df_redu_corr, color = colorRampPalette(c("#4979B6", "white", "#D73027"))(50),
                            labels_row = df_redu_corr$...1, 
                            border_color = NA, angle_col = 90)
pdf(file="/data1/wgs_APA/results/JW/redu_corr_heatmap.pdf", height = 11.5, width = 11.5)
p
dev.off()
```
